```python
y = re.totlist()
x = re.index.tolist()

plt.bar(x,y, allgn='center', alpha = 0.5)

plt.show()
```

```python
y = re.totlist()
x = re.index.tolist()

plt.bar(x,y, allgn='center', alpha = 0.7)
plt.title('울릉도 월별 풍속')
plt.clabel("월")
plt.ylabel('풍속')

for i in range(0, len(x)):
  plt.text(i, y[i], round(y[i], 1).ha = 'center')

plt.show()
```
