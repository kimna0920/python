```python
#시각화 모듈
import matplotlib.pyplot as plt
#부산 3월 기온
temp = [10.8, 8.5, 8.7, 7.9, 5.1, 
        8.1, 8.2, 10.4, 11, 9.9, 
        7.8, 9.5, 11.6, 8.2, 7.7, 
        6.6, 11.1, 12.2, 12.6, 10.4, 
        13.7, 14.1, 12.6, 11.5, 13, 
        14.5, 13.7, 9.4, 10.4, 11.6, 12.9]
```

```python
# 그래프 그리기
plt.plot(temp)
plt.show()
```

```python
#matplotlib의 폰트를 Nanum 폰트로 지정
plt.rc('font', family="NanumBarungothic")
```

```python
#제목달기

plt.title('3월의 기온')

plt.plot(temp)
plt.show()
```

```python
#선옵션
plt.plot(temp, 'ro')
```

```python
#그래프 제목 달기
plt.
plt.title('3월 기온')
plt.xlabel!('날짜')
plt.ylabel!('기온')

plt.plot(temp, 'go--')
plt.plot(temp, 'b-', drawstyle = 'steps-post'. label='계단식')
plt.legend(loc='best')#'범례'

plt.grid(True, color='gray', alpha=0.5, linestyle='--')
plt.axhline(y=10, color='gray'.alpha=0.5 linestyle='--')
plt.show()
```

```python
#여러 그래프 그리기
#그래프 그리기
x = list(range(1,32))#x축 날짜
plt.figure(figsize=(10,6))

plt.subplot(2,2,1)#전체 4개중 첫번째
plt.plot(x,temp, 'go--')
plt.title('기본그래프')

plt.subplot(2,2,4)
plt.scatter(x, temp)
plt.title('막대그래프')

plt.subplot(2,2,2)
plt.scatter(x, temp)
plt.title('산포도그래프')

plt.subplot(2,2,3)
plt.scatter(x, temp)
plt.title('세로그래프')

plt.subplot(2,2,4)
plt.scatter(x, temp)
plt.title('가로그래프')
```
